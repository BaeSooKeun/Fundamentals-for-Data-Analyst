### 관계 시각화
#### 산점도를 그릴 때는 극단치를 제거하고서 그리는 것이 좋다. 극단치로 인해 주요 분포 구간이 압축되어 시각화의 효율이 떨어지기 때문
#### 만일, 데이터가 너무 많아서 점들이 서로 겹쳐 정보를 파악하기 어렵다면 농도나 색상을 다르게 하여 표현
#### 산점도는 두개의 변수 간 관계만 표현할 수 있다는 단점이 있다.
#### 버블차트를 이용하면 세 가지 요소의 상관관계를 표현할 수 있다.
#### 버블차트 해석 시 주의사항: 지름이 아닌 면적으로 판단. because 지름이 2배= 면적이 4배

```python
# 필요한 패키지 설치

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
plt.rcParams['figure.dpi'] = 300
```

```python
# 데이터 불러오기
df = pd.read_csv("C:/Users/qorud/OneDrive/바탕 화면/Github/Data Analysis Method/50_Startups.csv")

# 데이터 샘플 확인
df.head()
```

```python
# 기본 산점도 시각화

plt.scatter(df['R&D Spend'], df['Profit'], s = 50, alpha = 0.4)
plt.show()
```

#### 기본 산점도를 시각화하기 위해 R&D 비용을 X축, 수익을 Y축으로 설정했음
#### 점 크기(s=50), 투명도(alpha=0.4) 로 설정하여 점이 겹치는 부분을 효과적으로 확인 가능
#### R&D 비용이 증가할수록 수익도 선형적으로 증가하는 것을 확인가능

```python
# 산점도에 회귀선 추가
ax = sns.lmplot(x='R&D Spend', y='Profit', data= df)
```

```python
# 네 가지 요소의 정보를 포함한 산점도 시각화

plt.scatter(df['R&D Spend'], df['Profit'], s=df['Marketing Spend']*0.001, 
            c=df['Administration'], alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

#### 마지막으로 마케팅 비용과 관리 비용 칼럼의 요소를 추가하여 버블 차트 형태로 시각화. 마케팅 비용: 버블의 크기, 관리 비용:색상
#### 이에 따라 마케팅 비용과 관리 비용이 어느 정도의 상관관계가 있다는 것을 확인할 수 있음.
