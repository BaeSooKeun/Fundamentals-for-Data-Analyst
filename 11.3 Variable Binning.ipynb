### 변수 구간화
#### 이산형 변수를 범주형 변수로 비즈니스적 상황에 맞도록 변환시킴으로써 데이터의 해석이나 예측, 분류 모델을 의도에 맞도록 유도할 수 있는 것이다.
#### 변수가 종속변수를 제대로 설명할 수 있도록 구간화가 잘되면 IV값(Information Value)이 높아지는 것이다

```python
# 필요한 패키지 설치
!pip install xverse
from xverse.transformer import WOE
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

plt.rcParams['figure.dpi'] = 300
```

```python
# 데이터 불러오기
df = pd.read_csv("C:/Users/qorud/OneDrive/바탕 화면/Github/Data Analysis Method/heart_2020_cleaned.csv")

# 데이터 샘플 확인
df.head()

# 이번 실습에선 BMI, PhysicalHealth, PhysicalHealth를 사용한다.
```

```python
# BMI 컬럼의 분포 확인

df['BMI'].describe()

# 아래 결과와 같이 12~94 사이의 값을 가지고 있는 것을 확인할 수 있다.
```

```python
# BMI 컬럼 분포 시각화

%matplotlib inline
sns.displot(df['BMI'],height = 5, aspect = 3)

# BMI의 분포는 대부분 20~40 사이에 위치해 있고 60 이상부터는 관측치가 희소하다.
```

```python
# 임의로 단순 구간화

df1 = df.copy() # 데이터셋 복사

# 구간화용 빈 컬럼 생성 - 생략해도 되지만 바로 옆에 붙여 보기 위함
df1.insert(2, 'BMI_bin', 0) 

df1.loc[df1['BMI'] <= 20, 'BMI_bin'] = 'a'
df1.loc[(df1['BMI'] > 20) & (df1['BMI'] <= 30), 'BMI_bin'] = 'b'
df1.loc[(df1['BMI'] > 30) & (df1['BMI'] <= 40), 'BMI_bin'] = 'c'
df1.loc[(df1['BMI'] > 40) & (df1['BMI'] <= 50), 'BMI_bin'] = 'd'
df1.loc[(df1['BMI'] > 50) & (df1['BMI'] <= 60), 'BMI_bin'] = 'e'
df1.loc[(df1['BMI'] > 60) & (df1['BMI'] <= 70), 'BMI_bin'] = 'f'
df1.loc[df1['BMI'] > 70, 'BMI_bin'] = 'g'

df1.head()
```

```python
# 구간화 변수 분포 시각화

sns.displot(df1['BMI_bin'],height = 5, aspect = 3)

# 다음과 같이 7개의 범주로 구분하여 분포를 단순화했다.
```

```python
#  cut() 함수 사용하여 임의로 구간화
df1.insert(3, 'BMI_bin2', 0) # 구간화용 빈 컬럼 생성 

df1['BMI_bin2'] = pd.cut(df1.BMI, bins=[0, 20, 30, 40, 50, 60, 70, 95]
                         , labels=['a', 'b', 'c', 'd', 'e', 'f', 'g'])

df1.head()
```

```python
# BMI_bin2 구간 별 관측치 수 집계

df1.BMI_bin2.value_counts().to_frame().style.background_gradient(cmap='winter')

# 구간별 관측ㅊ 수 집계를 확인해 보면 BMI가 20~30에 해당하는 B 구간이 202,548개로 가장 많고 70~95에 해당하는 g 구간이 129개로 가장 적다
```

```python
# qcut() 함수 사용하여 자동 구간화
df1.insert(4, 'BMI_bin3', 0) # 구간화용 빈 컬럼 생성 

df1['BMI_bin3'] = pd.qcut(df1.BMI, q=7, labels=['a', 'b', 'c', 'd', 'e', 'f', 'g'])

df1.head()
```

```python
# BMI_bin3 구간 별 관측치 수 집계

df1.BMI_bin3.value_counts().to_frame().style.background_gradient(cmap='winter')

#qcut 함수를 사용하여 자동으로 구간화를 적용했다. 
#관측치 수가 유사한 수준으로 배정되었고 BMI 칼럼과 같이 긴 꼬리 분포를 가진 경우에는 동일한 간격으로 구간화를 할 경우 포함되는 관측치가 너무 희소한 범주가 생겨 모델의 성능을 감소시킬 수 있다.
# 이에 다음과 같은 구간화 방식이 유용하게 사용될 수 있다.
```

```python
# BMI_bin3 분포 시각화

sns.displot(df1['BMI_bin3'],height = 5, aspect = 3)
```

```python
# WOE를 사용한 변수 구간화

df2 = df.copy()  # 데이터셋 복사


# xverse 함수 적용을 위한 더미변수 변환
df2=pd.get_dummies(df)

# 구간화 할 컬럼(X), 기준 컬럼(y) 지정
X = df2[['PhysicalHealth']]
y = df2[['KidneyDisease_Yes']]

y = y.T.squeeze() # 차원 축소

# WOE 모델 설정 및 적용
clf = WOE()
clf.fit(X, y)

# 구간 기준점 및 eight of Evidence 값 테이블 생성
a=clf.woe_df 

#Information Value 데이블 생성
b=clf.iv_df

a.head()
```
